```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
fmri =sns.load_dataset('fmri')
fmri.head()
```

```python
# relplot 
sns.relplot(x='subject', y='signal', data= fmri)
```

```python
# catplot
sns.catplot(x="region", y="signal", hue="event", data=fmri, kind='boxen')
```

```python
fmri.describe()
```

```python
sns.catplot(x='timepoint', y='signal', data= fmri, kind='box')
```

```python
# lineplot
sns.lineplot(x='subject', y='signal',data=fmri)
```
